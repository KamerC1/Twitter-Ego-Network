Dizionario che contiene i path delle activeEgoNetwork:
- key -> periodo temporale (es. 2020-01-01-2020-01-02)
- value -> path della activeEgoNetwork

```python
from src.dataProcessing.slidingWindow import get_period_list, utc_to_string
from src.utils.globalVariable import REGULAR_ACTIVE_USER_DIR


activeEgoNetwork_path_dict = {}
period_list = get_period_list()[-3:]

for first_period, second_period in period_list:
    first_period_utc = utc_to_string(first_period)
    second_period_utc = utc_to_string(second_period)

    activeEgoNetwork_dir = f"{PATH}/{REGULAR_ACTIVE_USER_DIR}/{first_period_utc}-{second_period_utc}/activeEgoNetwork"
    activeEgoNetwork_path_dict[f"{first_period_utc}-{second_period_utc}"] = activeEgoNetwork_dir
```

Meanshift e jenks

```python

from src.dataProcessing.clustering.Clustering import Clustering
from src.dataProcessing.clustering.JenksClustering import JenksClustering


for key, value in activeEgoNetwork_path_dict.items():
       SAVE_PATH = f"{PATH}/{REGULAR_ACTIVE_USER_DIR}/{key}/clustering"
       contactFrequencyClustering = Clustering(value)
       contactFrequencyClustering.save_clustering(f"{SAVE_PATH}/meanshift")

       jenksClustering = JenksClustering(value, f"{SAVE_PATH}/meanshift")
       jenksClustering.save_clustering(f"{SAVE_PATH}/jenks")
```
